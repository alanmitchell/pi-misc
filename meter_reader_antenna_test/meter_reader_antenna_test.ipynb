```python
import pandas as pd
```

```python
def analyze(fname):
    print(fname)
    df = pd.read_csv(fname,header=None)
    print(df.iloc[:, 3].value_counts())
    print('Total Readings:', len(df))
    print(df.iloc[0])
    print(df.iloc[-1])
    print()
```

```python
analyze('ebay_ant.csv')     # 18 minutes of data, 50 readings, 2.8 rds/min
analyze('ebay_ant2.csv')     # 25 minutes of data, 66 readings, 2.6 rds/min
analyze('stock_ant.csv')     # 20 minutes of data, 55 readings, 2.8 rds/min
analyze('stock_ant2.csv')     #  26 minutes of data, 70 readings, 2.7 rds/min
analyze('stock_extended_ant.csv')  # 27.5 minutes, 46 readings, 1.7 rds/min
```
